```python
import numpy as np
import pandas as pd
```

****Let us Read The URL Through Pandas****

```python
url = 'https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'
dfs = pd.read_html(url)

#This is to check how many relevant lists are available in the epage
print("The number of lists is:",len(dfs))
```

****There are 3 lists on the page. Lets us start with the first one****

```python
#We use '0' with the dataframe to check whether this is th elist we require
df = dfs[0]

print("Voila! You One Lucky Champ, This is the List You Were Looking For!")
df.head(12)
```

****"Now We Need To Drop The Bouroughs with Values 'Not Assigned'"****

To do this we use dataframe.drop function and also we need to specify the values to be dropped in the specific columns

```python
#Let us specify the Values we need to drop
indexNames = df[ df['Borough'] == "Not assigned" ].index

#Let us use the 'drop' funtion
df.drop(indexNames , inplace=True)

print("Boy o Boy, You are Right on Target, Boom-Ley!")
df.head(12)
```

****But There is an Issue, the Index is starting from 2 and since the rows have been dropped so has the index value
Now Let us Set the index value back****

```python
#Checking the shpae to get the exact number of rows now
print("Your Wish is My Command Boss!",df.shape)
```

Now let us Use the pd.RangeIndex to start and stop the index vlaues

```python
df.index = pd.RangeIndex(start=0, stop=103, step=1)
```

```python
print("Neat & Clean as You Like Boss!")

df.head(11)
```
