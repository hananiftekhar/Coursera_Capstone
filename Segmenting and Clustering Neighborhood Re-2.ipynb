```python
import numpy as np
import pandas as pd
```

****Let us Read The URL Through Pandas****

```python
url = 'https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'
dfs = pd.read_html(url)

#This is to check how many relevant lists are available in the epage
print("The number of lists is:",len(dfs))
```

```python
#We use '0' with the dataframe to check whether this is th elist we require
df = dfs[0]

print("Voila! You One Lucky Champ, This is the List You Were Looking For!")
df.head(12)
```

```python
#Let us specify the Values we need to drop
indexNames = df[ df['Borough'] == "Not assigned" ].index

#Let us use the 'drop' funtion
df.drop(indexNames , inplace=True)
```

```python
#Now let us Use the pd.RangeIndex to start and stop the index vlaues
df.index = pd.RangeIndex(start=0, stop=103, step=1)
```

```python
df.head(11)
```

Let us Import the csv file for latitude and longitude

```python
toronto_geocsv = 'https://cocl.us/Geospatial_data'
!wget -q -O 'toronto_m.geospatial_data.csv' toronto_geocsv

#settign index to 'postal code'
geocsv_data = pd.read_csv(toronto_geocsv).set_index("Postal Code")
geocsv_data.head()
```

Now let us comine both dataframes 

```python
df_com = pd.merge(geocsv_data, df , on = 'Postal Code')
```

```python
df_com.head()
```

No Let us Rearrange the Columns accordign to Requirement with Lat & Long as the last two columns

```python
df_com = df_com[["Postal Code", "Borough", "Neighborhood", "Latitude", "Longitude"]]
print("There You Have it Champ!")
df_com.head(12)
```
